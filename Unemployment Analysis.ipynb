### Visualizing the Unemployment Rate

```python
sns.scatterplot(x=' Estimated Unemployment Rate (%)', y=' Estimated Employed', data=df)
```

```python
plt.plot(df[' Estimated Unemployment Rate (%)'], df['Region'])
plt.xlabel('Unemployment Rate')
plt.ylabel('region')
plt.title('Estimation of Unemployment rate over Region')
plt.show()
```

```python
plt.plot(df[' Estimated Unemployment Rate (%)'], df[' Date'])
plt.xlabel('Unemployment Rate')
plt.ylabel('Year')
plt.title('Estimation of Unemployment rate over Time')
plt.show()
```

```python
plt.plot(df[' Estimated Unemployment Rate (%)'], df[' Frequency'])
plt.xlabel('Unemployment Rate')
plt.ylabel('Frequency')
plt.title('Occurence of Unemployment')
plt.show()
```

```python
plt.plot(df[' Estimated Unemployment Rate (%)'], df[' Estimated Labour Participation Rate (%)'])
plt.xlabel('Unemployment Rate')
plt.ylabel('Employment rate')
plt.title('Estimation of Unemployment Rate Over Employment rate')
plt.show()
```

```python
plt.plot(df[' Estimated Unemployment Rate (%)'], df['Region.1'])
plt.xlabel('Unemployment Rate')
plt.ylabel('Directions')
plt.title('Estimation of Unemployment rate through Directions')
plt.show()
```

```python
plt.plot(df[' Estimated Unemployment Rate (%)'], df['longitude'])
plt.xlabel('Unemployment Rate')
plt.ylabel('Longitude')
plt.title('Estimation of Unemployment rate through Longitude Axis')
plt.show()
```

```python
plt.plot(df[' Estimated Unemployment Rate (%)'], df['latitude'])
plt.xlabel('Unemployment Rate')
plt.ylabel('Latitude')
plt.title('Estimation of Unemployment rate through Latitude Axis')
plt.show()
```

### Analyzing the Unemployment Rate

```python
# Calculate the Average Unemployment rate
average_rate = df[' Estimated Unemployment Rate (%)'].mean()
print('Average Unemployment Rate:', average_rate)
```

```python
# Calculate the Minimum Unemployment rate
min_rate = df[' Estimated Unemployment Rate (%)'].min()
print('Minimum Unemployment Rate:', min_rate)
```

```python
# Calculate the Maximum Unemployment rate
max_rate = df[' Estimated Unemployment Rate (%)'].max()
print('Maximum Unemployment Rate:', max_rate)
```

```python
# Calculate the Total number of Unemployed people
total_unemployed = df[' Estimated Unemployment Rate (%)'].sum()
print('Total Unemployed:', total_unemployed)
```
